Q1. Create a function in python to read text file and replace specific content of file
    File Name- example.txt
    origin file content- This is placement assignment
    Replace string - Placement should be replaced with screening
    Replaced file content - This is screening assignment

```python
## Function to read file
def read_file():
    try:
        ##read file example.txt
        file=open('example.txt')
        data=file.read()
        
        ##Split data of string into list elements
        splitted_data= data.split(" ")
        final_string=' '
        splitted_lower_data=[]
        
        ##loop through spliited list to find word placement
        for i in range(len(splitted_data)):
            
            ## convert data into lower case top compare the string
            splitted_lower_data.append(splitted_data[i].lower())
            f=0
            
            ##if word placement is found replace that list element with screening
            if(splitted_lower_data[i]=="placement"):
                splitted_data[i]='screening'
                f=1
                break
        
        ##Join the splitted elements in list to form a single string
        final_string=final_string.join(splitted_data)
        
        ##if word placement is not there in file then handlit with flag
        if(f==0):
            print(" No String Matching with word Placement")
        
        ## close the file after use
        file.close()
        
        ##return the final string
        return final_string
    
    except(e):
        print("Exception Raised",e)
        
## Function to write file
## It takes input of updated string 
def write_file(Updated_string):
    
    ## Open the file in write mode
    file=open('example.txt','w')
    
    ##Write the string in file
    file.writelines(Updated_string)

## Function call:Read File
New_string=read_file()

## Function call:Write file
write_file(New_string)
```

Abstract Class

```python
## importing required modules
import math
from abc import abstractmethod

## Calculation class which can calculate Area,Perimeter of perticular shape
class Calculation():
    
    ## abstract method which has only declaration but no definition
    @abstractmethod
    def Area(self):
        pass
    
    @abstractmethod
    def perimeter(self):
        pass
    
    ## non abstarct class which has declaration and definition
    def functionality(self):
        print("This class is common for all the shapes.")

## Circle class inheriting Calculation Class
class Circle(Calculation):
    
    ## defining existing abstract method according to shape 
    def Area(self,radius):
        return (3.14*(radius**2))
    
    ## defining existing abstract method according to shape
    def perimeter(self,radius):
        return (2*3.14*radius)
    
    ## This is own method of Circle class
    def shape(self):
        print("This is a circle")

## Rectangle class inheriting Calculation Class
class Rectangle(Calculation):
    
    ## defining existing abstract method according to shape 
    def Area(self,length,breadth):
        return (length*breadth)
    
    ## defining existing abstract method according to shape 
    def perimeter(self,length,breadth):
        return ((2*length)+(2*breadth))

## Octagon class inheriting Calculation Class
class Octagon(Calculation):
    
    ## defining existing abstract method according to shape 
    def Area(self,Side):
        return (2*Side*Side*(1+math.sqrt(2)))
    
    ## defining existing abstract method according to shape 
    def perimeter(self,Side):
        return 8*Side
    
    ## This is own method of Octagon Class
    def sides(self):
        print("Octagon has 8 Sides")

## Object Creation for Circle Class
circle=Circle()

##function call
print("Area of Circle=",circle.Area(10))
circle.shape()

## Object creation for Octagon class
octagon=Octagon()

## Function call
print("Perimeter of Octagon=",octagon.perimeter(10))
octagon.functionality()
```

```python
## Furniture class- has 3 options 
## 1) Sofa 2) Cupboard 3)wardrobe as 3 different methods
class Furniture():
    
    ##Sofa Method with 2 variations type1 and type2
    def Sofa(self,cast):
        if(cast==1):
            return 7000
        if(cast==2):
            return 10000
    
    ##Cupboard Method with 3 variations type1 and type2 and type3
    def cupboard(self,cast):
        if(cast==1):
            return 10000
        if(cast==2):
            return 15000
        if(cast==3):
            return 22000
    
    ##Wardrobe Method with 2 variations type1 and type2
    def wardrobe(self,cast):
        if(cast==1):
            return 8000
        if(cast==2):
            return 10000

## Doors class- has 2 options 
## 1) glass_doors 2) wooden_doors  as 2 different methods
class Doors():
    def glass_doors(self,n_doors):
        return n_doors*3000
    
    def wooden_doors(self,n_doors):
        return n_doors*5000

## Fridge class- has 2 options 
## 1) Whirlpool 2) LG  as 2 different methods
class Fridge():
    
    ##Whirlpool Method with 2 variations double doors and triple doors
    def whirlpool(self,cast):
        if(cast==1):
            return 15000
        if(cast==2):
            return 20000
    
    ##LG Method with 2 variations double doors and triple doors
    def LG(self,cast):
        if(cast==1):
            return 18000
        if(cast==2):
            return 25000

## Wifi_setup class- has 2 options 
## 1) hathway 2) Jio  as 2 different methods
class Wifi_setup():
    
    ##hathway Method with 3 variations 30mbps and 50mbps and 100mbps
    def hathway(self,cast):
        if(cast==1):
            return 600
        if(cast==2):
            return 900
        if(cast==3):
            return 1200
    
    ##jio Method with 2 variations 50mbps and 100mbps
    def jio(self,cast):
        if(cast==1):
            return 500
        if(cast==2):
            return 1000

## Main Class-New_house which is inheriting all Furniture,Doors,Fridge,Wifi_setup classes- Multiple Inheritance
class New_house(Furniture,Doors,Fridge,Wifi_setup):
    
    ## Calculate_total_cost method which returns total_cost of setting up house
    def Calculate_total_cost(self):
        c='1'
        
        ##base price of flat
        house_base_price=2000000
        
        ##Initialize cost to basic house price 
        cost=house_base_price
        
        ## Maintaining list of added products
        items_list=['Flat Basic Price']
        f=0
        
        ## Starting infinite loop to give user flexibility to keep on adding products
        print("For New House Setup you can add following-")
        while(c=='1'):
            print("1. Furniture\n2.Doors\n3.Fridge\n4.Wifi_setup")
            
            ##Taking requirement type among 1,2,3
            req_1=input()
            #Furniture
            if(req_1=='1'):
                ## Creating object of Furniture class
                furniture=Furniture()
                print("1.Sofa\n2.Cupboard \n3.Wardrobe")
                req_2=input()
                ##Sofa
                if(req_2=='1'):
                    print("1.Type1 Sofa\n2.Type2 Sofa")
                    req_3=input()
                    ## Type 1 sofa
                    if(req_3=='1'):
                        ## Adding selected element in list
                        items_list.append("Type1 Sofa")
                        ## adding cost of perticular item
                        cost+=furniture.Sofa(1)
                        
                    ##Type 2 sofa
                    if(req_3=='2'):
                        ## Adding selected element in list
                        items_list.append("Type2 Sofa")
                        ## adding cost of perticular item
                        cost+=furniture.Sofa(2)
                    else:
                        f=1

                #Cupboard
                if(req_2=='2'):
                    print("1.Type1 Cupborad-3 parts\n2.Type2 Cupboard-5 Parts\n3.Type3 Cupboard- 7 Parts")
                    req_3=input()
                    ## Type 1 Cupboard
                    if(req_3=='1'):
                        items_list.append("Type1 Cupborad-3 parts")
                        cost+=furniture.cupboard(1)
                        
                    ##Type 2 Cupboard
                    if(req_3=='2'):
                        items_list.append("Type2 Cupborad-5 parts")
                        cost+=furniture.cupboard(2)
                        
                    ## Type 3 Cupboard
                    if(req_3=='3'):
                        items_list.append("Type3 Cupborad-7 parts")
                        cost+=furniture.cupboard(3)
                    else:
                        f=1
                    
                #Wardrobe
                if(req_2=='3'):
                    print("1.Type1 wardrobe\n2.Type2 Wardrobe with mirror")
                    req_3=input()
                    if(req_3==1):
                        ##Type 1 wardrobe
                        items_list.append("Type1 wardrobe")
                        cost+=furniture.wardrobe(1)
                    elif(req_3=='2'):
                        ##Type 2 Cupboard
                        items_list.append("Type2 wardrobe")
                        cost+=furniture.wardrobe(2)
                    else:
                        f=1
                else:
                    f=1
            ##Doors
            if(req_1=='2'):
                #Creating object of door class
                doors=Doors()
                print("1.Glass Doors\n2.Wooden Doors ")
                req_2=input()
                ##Glass Doors
                if(req_2=='1'):
                    req_3=input("No of Doors")
                    items_list.append(req_3 + " Glass Doors")
                    cost+=doors.glass_doors(int(req_3))
                #Wooden Doors
                if(req_2=='2'):
                    req_3=input("No of Doors")
                    items_list.append(req_3 + " Wooden Doors")
                    cost+=doors.wooden_doors(int(req_3))
                else:
                    f=1
            ##Fridge
            if(req_1=='3'):
                fridge=Fridge()
                print("1.Whirlpool\n2.LG")
                req_2=input()
                ##Whirlpool
                if(req_2=='1'):
                    print("1.Whirlpool double door\n2.Whirlpool triple door")
                    req_3=input()
                    ## Double door whirlpool
                    if(req_3=='1'):
                        items_list.append("Whirlpool double door Fridge")
                        cost+=fridge.whirlpool(1)
                    ## Triple door whirlpool
                    if(req_3=='2'):
                        items_list.append("Whirlpool Triple door Fridge")
                        cost+=fridge.whirlpool(2)
                    else:
                        f=1
                #LG
                if(req_2=='2'):
                    print("1.LG  double door\n2.LG triple door")
                    req_3=input()
                    ##Double door LG
                    if(req_3=='1'):
                        items_list.append("LG double door Fridge")
                        cost+=fridge.LG(1)
                    ##Triple door LG
                    if(req_3=='2'):
                        items_list.append("LG triple door Fridge")
                        cost+=fridge.LG(2)
                    else:
                        f=1
                else:
                    f=1
            ## Wifi Setup
            if(req_1=='4'):
                wifi=Wifi_setup()
                print("1.hathway\n2.Jio")
                req_2=input()
                ##hathway
                if(req_2=='1'):
                    print("1.hathway 30 mbps\n2.hathway 50 mbps\n3.hathway 100 mbps")
                    req_3=input()
                    ## hathway 30mbps
                    if(req_3=='1'):
                        items_list.append("hathway Wifi 30mbps")
                        cost+=wifi.hathway(1)
                    ##hathway 50 mbps
                    if(req_3=='2'):
                        items_list.append("hathway Wifi 50mbps")
                        cost+=wifi.hathway(2)
                    ##hathway 100mbps
                    if(req_3=='3'):
                        items_list.append("hathway Wifi 100mbps")
                        cost+=wifi.hathway(3)
                    else:
                        f=1
                #Jio
                if(req_2=='2'):
                    print("1.Jio 50 mbps\n2.Jio 100 mbps")
                    req_3=input()
                    ## Jio 50mbps
                    if(req_3=='1'):
                        items_list.append("jio Wifi 50mbps")
                        cost+=wifi.jio(1)
                    ##jio 100mbps
                    if(req_3=='2'):
                        items_list.append("jio Wifi 100mbps")
                        cost+=wifi.jio(2)
                    else:
                        f=1
                else:
                    f=1
            else:
                if(cost==house_base_price):
                    print("Invalid option selected")
            ##Input to loop continuation
            c=input("press 1 to continue adding-")
        ##Priniting list of items added
        print("List of items added to setup house=",items_list)
        
        ##returning cost
        return cost
##Object creation of New_house
new_house=New_house()

##accepting cost as returned value
total_cost=new_house.Calculate_total_cost()

##printing total cost
print("Total Cost for setting up entire house=",total_cost)
```

Decorator

```python
## decorated Function
def factorial_decorator(fn,num):
    def decorated_function():
        
        ## calling factorial function and returning output of factorial to main function call
        return fn(num)
    
    # returns a function
    return decorated_function
 
## Factorial fucntion which calculates factorial and returns
def Factorial(n):
    fact=1
    for i in range(2,n+1):
        fact*=i
    ## returning factorial
    return fact
    #print("Factorial of",n," =",fact)

c='1'

##Initailting infinite loop to give user flexibility to keep on calculatiing factorial
while(True):
    if(c=='1'):
        
        ##Input of number to find factorial
        num=int(input("Enter number to find factorial:"))
        
        ## calling function-decorator and passing factorial function as argument along with input number
        factorial_dec = factorial_decorator(Factorial,num)
        
        ##Print Final output
        print(factorial_dec())
    else:
        
        ##break the loop if anything ither than 1 is entered
        break 
    
    ##Taking input to continue loop
    c=input("press 1 to continue-")

## Function call
factorial_dec()
```
